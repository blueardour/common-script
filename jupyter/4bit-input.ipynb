```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import sys, os
import matplotlib as mpl
mpl.rcParams['figure.dpi']= 200
```

```python
for i in os.walk('/home/chenp/Documents'):
    root, path, lists = i
    print(root)
    print(path)
    print(len(lists))
   
    for i, fp in enumerate(lists):
        name = os.path.join(root, fp)
        print(name)
        img = cv2.imread(name)
        img2 = img[:,:,::-1]
        img4 = img2 // 16 * 16
        img5 = img2 // 16
        #
        plt.figure()
        f, axarr = plt.subplots(1,3) 
        axarr[0].imshow(img2)
        axarr[1].imshow(img4)
        axarr[2].imshow(img5)
        #plt.show()
        
        if i > 3:
            break
                        
```
